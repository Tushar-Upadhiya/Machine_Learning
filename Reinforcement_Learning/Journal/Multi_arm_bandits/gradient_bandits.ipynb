```python
import math
import random
```

```python
# 3-armed Bandit Problem
alpha = 0.1
num_actions = 3
preferences = [0.0] * num_actions
baseline_reward = 0.0
step_count = 0
true_rewards = [1.5, 1.0, 3.0]  # True rewards for each action

print("Initial Preferences: ", preferences)
 
for step in range (1,11):
    step_count+=1
    max_pref=max(preferences)
    exp_prefs = [math.exp(h-max_pref) for h in preferences]
    sum_exp_prefs = sum(exp_prefs)
    probabilities = [exp_pref / sum_exp_prefs for exp_pref in exp_prefs]
    action = random.choices(range(num_actions),probabilities)[0]
    reward = random.normalvariate(true_rewards[action],1.0)
    if step_count == 1:
        baseline_reward = reward
    else:
        baseline_reward += (reward - baseline_reward) / step_count
    for a in range(num_actions):
        if a == action:
            preferences[a] += alpha * (reward - baseline_reward) * (1 - probabilities[a])
        else:
            preferences[a] -= alpha * (reward - baseline_reward) * probabilities[a]

    prob_strings = [f"{p:.2f}" for p in probabilities]
    pref_strings = [f"{h:.2f}" for h in preferences]
    print(f"Step {step}: Took Arm {action}, Reward: {reward:.2f}, Baseline: {baseline_reward:.2f}")
    print(f"        Action Probabilities: {prob_strings}")
    print(f"        Updated Preferences:  {pref_strings}\n")
```
