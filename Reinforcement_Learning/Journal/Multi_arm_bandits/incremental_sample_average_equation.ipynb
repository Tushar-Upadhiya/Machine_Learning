```python
import random
def step():
    return 5.0+ random.uniform(-2,2)

Q_k = 0
k = 0

print(f"Initial Setup: Q_1 = {Q_k:.2f}, Count = {k}\n")

for i in range (1,6):
    R_k = step()
    k+=1
    Q_next = Q_k + (1/k) * (R_k - Q_k)
    print(f"  Received Reward (R_k): {R_k:.2f}")
    print(f"  Old Estimate (Q_k):   {Q_k:.2f}")
    print(f"  Error (R_k - Q_k):     {(R_k - Q_k):.2f}")
    print(f"  New Estimate (Q_k+1): {Q_next:.2f}\n")
    Q_k = Q_next
```

```python
#lets code with constant step size alpha = 0.1
alpha = 0.1
goal = 50
estimate = 0.0
for step in range (1,100):
    reward = goal+random.uniform(-5,5)
    error = reward-estimate
    estimate += alpha * error
    print(f"Step {step}:")
    print(f"  Scale Readout (Reward): {reward:.2f}")
    print(f"  Our Surprise (Error):   {error:.2f}")
    print(f"  Updated Estimate:       {estimate:.2f}\n")
```
