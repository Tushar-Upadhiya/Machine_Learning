```python
import random
import math
```

```python
q_value = 4.2
total_steps = 100
action_steps=5
c = 2

uncertainty_parameter = math.sqrt(math.log(total_steps)/action_steps)
ucb_value = q_value + c * uncertainty_parameter
print(f"Exploitation Value (Q): {q_value}")
print(f"Exploration Bonus:       {uncertainty_parameter:.2f}")
print(f"Total UCB Value:        {ucb_value:.2f}")
```
